## Sinusovy signal


```python
import matplotlib.pyplot as plt
import numpy as np


fs = 44100
f = 440
t = np.linspace(0, 1, fs, endpoint=False)

y = np.sin(2 * np.pi * f * t)

plt.figure(figsize=(10, 4))
plt.plot(t[:1000], y[:1000])
plt.title("440 Hz Sine Wave")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

## Trojuhelnikovy signal


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal


fs = 44100  # Sampling rate (Hz)
f = 440  # Frequency of the saw wave (Hz)
t = np.linspace(0, 1, fs, endpoint=False)  # 1 second of signal

# Generate a sawtooth wave (goes from -1 to +1)
y = signal.sawtooth(2 * np.pi * f * t)

plt.figure(figsize=(10, 4))
plt.plot(t[:1000], y[:1000])  # only show first 1000 samples
plt.title("440 Hz Sawtooth Wave")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
